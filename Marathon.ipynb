```python
import csv
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
data = pd.read_csv('marathon_results_2015.csv')
#data.head()
```

```python
#clean data for period pace analysis
index = ['Name','M/F','5K','10K','15K','20K','Half','25K','30K','35K','40K'] #we can add any column here to do more analysis
time = data[index]
for i in index:
    time = time[ ~ time[i].str.contains('-')] #remove rows contained ''-''
time
```

```python
#change time string to secs
def str2secs(s):
    h,m,s = map(float,s.split(":"))
    return h*3600+m*60+s
```

```python
#create new columns for time in secs
index1 = ['5K','10K','15K','20K','Half','25K','30K','35K','40K']
index2 = ['5K_secs','10K_secs','15K_secs','20K_secs','Half_secs','25K_secs','30K_secs','35K_secs','40K_secs']
period = ['period_1','period_2','period_3','period_4','period_5','period_6','period_7','period_8']
for i in range(0,len(index1)):
    new_time = list(time[index1[i]])
    time[index2[i]] = list(map(str2secs,new_time)) 
for j in range(0,8):
    time[period[j]] = time[index2[j+1]] - time[index2[j]]
time.head()
```
