```python
import csv
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
data2015=pd.read_csv("marathon_results_2015.csv")
data2016=pd.read_csv("marathon_results_2016.csv")
data2017=pd.read_csv("marathon_results_2017.csv")
```

```python
len(data2015),len(data2016), len(data2017)
```

```python
data2016.sample(10)
```

```python
dataFemale2015=data2015[data2015['M/F']=='F']
dataFemale2016=data2016[data2016['M/F']=='F']
dataFemale2017=data2017[data2017['M/F']=='F']
dataMale2015=data2015[data2015['M/F']=='M']
dataMale2016=data2016[data2016['M/F']=='M']
dataMale2017=data2017[data2017['M/F']=='M']
```

```python
topCountries15= data2015['Country'].value_counts()
topCountries16= data2016['Country'].value_counts()
topCountries17= data2017['Country'].value_counts()
```

```python
topCountries15Female= dataFemale2015['Country'].value_counts()
topCountries15Male= dataMale2015['Country'].value_counts()
```

```python
plt.figure()
plt.subplot(111)
topCountries15Female[2:10].plot(kind='barh',alpha=0.5)
topCountries15Male[2:10].plot(kind='barh')
plt.show()
```

```python
topCountries15.index
```

```python
femaleChart = [sum(dataFemale2015['Country']==x) for x in topCountries15.index][2:]
```

```python
maleChart = [sum(dataMale2015['Country']==x) for x in topCountries15.index][2:]
```

```python
y = list(range(len(femaleChart)))
fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(10,15))

axes[0].barh(y, femaleChart, align='center', color='red', alpha=0.8)
axes[1].barh(y, maleChart, align='center', color='blue', alpha=0.8)
axes[0].invert_xaxis()
axes[0].set(yticks=y, yticklabels=topCountries15.index[2:])
axes[0].yaxis.tick_right()
plt.show()
```
